```python
root_folder = 'D:\\Research\\STM Tip Prep Data\\Gold 111 Data\\'
active_folder = '2016-02-13 Au(111) more STM and STS\\'
filename2 = 'gold 111007.sxm'
path2 = root_folder + active_folder + filename2
data = pySPM.SXM(path2)
data.list_channels()
data.header
```

# Data Pre-Processing

## Z Height Correction

```python
topo1 = data.get_channel('Z')
topo2 = data.get_channel('Z').correct_lines(inline=False)
topo3 = data.get_channel('Z').correct_plane(inline=False)

fig, ax = plt.subplots(1,3,figsize=(21, 7))
topo1.show(ax=ax[0])
ax[0].set_title("Original image")
topo2.show(ax=ax[1])
ax[1].set_title("Corrected by lines")
topo3.show(ax=ax[2])
ax[2].set_title("Corrected by slope");
```

# Fourier Transform of STM Height Images

## Au(111) Herringbone Surface Reconstruction

```python
herringbone_ft = np.fft.fft2(topo3.pixels)
herringbone_ft_ft_shifted = np.fft.fftshift(herringbone_ft)
herringbone_ft_magnitude_spectrum = np.log(np.abs(herringbone_ft_ft_shifted))
```

```python
fig, ax = plt.subplots(1,2,figsize=(21, 7))
plt.subplot(121), plt.imshow(topo3.pixels)
plt.subplot(122), plt.imshow(herringbone_ft_magnitude_spectrum)
```

```python
data_ft = np.fft.fft2(topo1.pixels)
data_ft_shifted = np.fft.fftshift(data_ft)
magnitude_spectrum = np.log(np.abs(data_ft_shifted))
fig, ax = plt.subplots(1,2,figsize=(21, 7))
plt.subplot(121), plt.imshow(topo1.pixels)
plt.subplot(122), plt.imshow(magnitude_spectrum)
```

## Au(111) NOT Herringbone

```python
filename1 = 'STM_Au(111)_TipConditioning_LN2_022.sxm'
path1 = root_folder + filename1
data2 = pySPM.SXM(path1)
data2.header
```

```python
lattice = data2.get_channel('Z').correct_plane(inline=False)
lattice_ft = np.fft.fft2(lattice.pixels)
lattice_ft_shifted = np.fft.fftshift(lattice_ft)
lattice_magnitude_spectrum = np.log(np.abs(lattice_ft_shifted))
fig, ax = plt.subplots(1,2,figsize=(21, 7))
plt.subplot(121), plt.imshow(lattice.pixels)
plt.subplot(122), plt.imshow(lattice_magnitude_spectrum)
```

```python
x = 1/np.abs(lattice_ft_shifted)
x[x < 5E-10]
```

```python
inverse_lattice_magnitude_spectrum = np.log(1/np.abs(lattice_ft_shifted))
fig, ax = plt.subplots(1,2,figsize=(21, 7))
plt.subplot(121), plt.imshow(lattice.pixels)
plt.subplot(122), plt.imshow(inverse_lattice_magnitude_spectrum)
```

## Fourier Transorm of Line Scans

```python
herringbone_line_scan_fft = np.abs(list(map(np.fft.fft, topo3.pixels)))
plt.plot(herringbone_line_scan_fft[250])
```

```python
step = 1.000E-8/512.0
n = herringbone_line_scan_fft[250].size
freq = np.fft.fftfreq(n, step)
plt.plot(freq, herringbone_line_scan_fft[250])
```

```python
lattice_line_scan_fft = np.abs(list(map(np.fft.fft, lattice.pixels)))
plt.plot(lattice_line_scan_fft[150])
```

```python
step = 4.000000E-9 / 256.0
n = lattice_line_scan_fft[150].size
freq = np.fft.fftfreq(n, step)
plt.plot(freq, lattice_line_scan_fft[150])
```
